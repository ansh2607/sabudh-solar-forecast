```python
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sb
import warnings as wn
wn.filterwarnings('ignore')

import matplotlib.pylab as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20, 6
plt.style.use('ggplot')
```

```python
yield_data=pd.read_csv('Combined_Solar_Dataset.csv')
events_data=pd.read_csv('final_Event_Dataset.csv')
weather_data=pd.read_csv('weather_data.csv')
```

```python
yield_data['Date']=pd.to_datetime(yield_data['Date'])
yield_data=yield_data[yield_data.Date.between('2019-02-01','2021-05-31')]
yield_data=yield_data.drop(columns=['Inverter 1 [kWh]'])
yield_data
```

```python
events_data['Date']=pd.to_datetime(events_data['Date'])
events_data=events_data[events_data.Date.between('2019-02-01','2021-05-31')]
events_data
```

```python
finalData=pd.merge(yield_data, events_data, on="Date",how='outer')
finalData=finalData.fillna(0)
```

```python
weather_data['Date']=pd.to_datetime(weather_data['Date'])
weather_data=weather_data[weather_data.Date.between('2019-02-01','2021-05-31')]
```

```python
finalData=pd.merge(finalData, weather_data, on='Date',how='outer')
```
