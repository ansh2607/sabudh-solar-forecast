```python
import pandas as pd
import numpy as np
import glob
import warnings as wn
import datetime as dt
```

```python
yield_filenames=glob.glob(r'Solar_Power_Prediction'+"/*.xls")
alarm_filenames18=glob.glob(r'Alarm History/2018'+"/*.xlsx")
alarm_filenames19=glob.glob(r'Alarm History/2019'+"/*.xlsx")
alarm_filenames20=glob.glob(r'Alarm History/2020'+"/*.xlsx")
alarm_filenames21=glob.glob(r'Alarm History/2021'+"/*.xlsx")
```

```python
def form_data(dataset,filenames):    
    for file in filenames:
        if filenames==yield_filenames:
            df=pd.concat(pd.read_excel(file,sheet_name=None,skiprows=16))
        else:
            df=pd.concat(pd.read_excel(file,sheet_name=None))
        dataset=dataset.append(df,ignore_index=True)
    dataset=dataset.sort_values(by='Date')
    return dataset
yield_dataset=pd.DataFrame()
yield_dataset=form_data(yield_dataset,yield_filenames)
#dataset['Date']=dataset['Date'].dt.strftime('%d/%m/%Y')
yield_dataset.head(25)
```

```python
alarm_dataset18=pd.DataFrame()
alarm_dataset19=pd.DataFrame()
alarm_dataset20=pd.DataFrame()
alarm_dataset21=pd.DataFrame()
#var_holder={}
#for i in range(18,22):
#    i=str(i)
alarm_dataset18=form_data(alarm_dataset18,alarm_filenames18)
alarm_dataset19=form_data(alarm_dataset19,alarm_filenames19)
alarm_dataset20=form_data(alarm_dataset20,alarm_filenames20)
alarm_dataset21=form_data(alarm_dataset21,alarm_filenames21)
alarm_dataset=pd.concat([alarm_dataset18,alarm_dataset19,alarm_dataset20,alarm_dataset21])
alarm_dataset.head(7)
```

```python
yield_dataset.to_csv('./Combined_Solar_Dataset.csv',index=False)
alarm_dataset.to_csv('./Combined_Alarm_Dataset.csv',index=False)
```
