```python
import pandas as pd
```

```python
weather1=pd.read_csv('history_data1.csv')
weather2=pd.read_csv('history_data2.csv')
weather3=pd.read_csv('history_data3.csv')
weather4=pd.read_csv('history_data4.csv')
weather1
```

```python
weather_data=pd.concat([weather1,weather2,weather3,weather4])
weather_data
```

```python
weather_data.columns
```

```python
weather_data.info()
```

```python
weather_data=weather_data.drop(columns=['Name','Wind Chill', 'Heat Index','Snow',
       'Snow Depth', 'Wind Speed', 'Wind Direction', 'Wind Gust', 'Visibility','Relative Humidity'])
```

```python
weather_data=weather_data.rename(columns={'Date time':'Date','Maximum Temperature':'Max','Minimum Temperature':'Min','Temperature':'Temp','Precipitation':'Prcp'})
weather_data
```

```python
weather_data['Date']=pd.to_datetime(weather_data['Date'])
#weather_data['Date']=weather_data['Date'].dt.strftime('%Y-%d-%m')
weather_data['Date']=weather_data['Date'].dt.strftime('%Y-%m-%d')
weather_data.info()
```

```python
weather_data
```

```python
weather_data.to_csv('./weather_data.csv',index=False)
```
